# YELP Checkin Dataset and its Metadata

```python
# checkin data
checkin_df = pd.read_json('/kaggle/input/yelp-dataset/yelp_academic_dataset_checkin.json', lines=True, nrows=5000)
checkin_df['date'] = checkin_df['date'].str.split(', ')
checkin_df = checkin_df.explode('date')
checkin_df['date'] = pd.to_datetime(checkin_df['date'], errors='coerce')
checkin_df = checkin_df.dropna(subset=['date'])
checkin_df['checkin_id'] = range(1, len(checkin_df) + 1)
# time period and scope of checkin
e_dt = str(checkin_df['date'].min())[:19]
l_dt = checkin_df['date'].max()
print("Checkin table earliest date:", e_dt)
print("Checkin table latest date:", l_dt)
# structure of checkin
checkin_df.info()
checkin_df.head()
```

This Checkin dataset provides timestamps for user visits to businesses. It has the following columns:

1. business_id: 22 character string which maps to business in business.json.
2. date: records the date in the YYYY-MM-DD HH:MM:SS format, and is a comma-separated list of timestamps for each checkin.

# YELP Business Dataset and its Metadata

```python
# business data
business_df = pd.read_json('/kaggle/input/yelp-dataset/yelp_academic_dataset_business.json', lines=True, nrows=5000)
business_df['attributes'] = business_df['attributes'].astype(str)
business_df['hours'] = business_df['hours'].astype(str)
# structure of business
business_df.info()
business_df.head()
```

This Business dataset Contains information on various businesses listed on Yelp, such as the business name, category, location (address, city, state, postal code, latitude/longitude), hours of operation, and star ratings. It has the following columns:

1. business_id: 22 character string for business id, maps to business in business.json
2. name : string for the business's name
3. address : string displaying the full address of the business
4. city : one of 8 metripolitan cities (string) in the US and Canada  
5. state: string for the state in which the city exists
6. postal code: string for postal code of the city
7. latitude: a float datatype for latitude of the city
8. longitude: a float datatype for longitude of the city
9. stars: a float showing rating in terms of stars
10. review-count: integer value for number of reviews obtained
11. is_open : integer, 0 or 1 for closed or open
12. attributes : business attributes with corresponding values. Some are objects. For instance: Business Parking (with 'garage', 'valet' etc)
13. categories : array of strings of business categories
14. hours: object of key day to value hours, hours are using a 24hr clock

# Business Queries for Opening a New Restaurant

# 1. Top 5 Most Popular Restaurant Categories in a City

```python
query = """
SELECT TOP 5 
    categories, 
    COUNT(business_id) AS num_restaurants
FROM BUSINESS
WHERE city = 'Philadelphia' AND categories LIKE '%Restaurants%'
GROUP BY categories
ORDER BY num_restaurants DESC;
"""

# Execute the query and store the results in a DataFrame
df_popular_categories = pd.read_sql(query, engine)
print(df_popular_categories)
```

**Insight:** This query reveals which types of restaurants are most common, indicating strong demand. If a certain category (like Italian or Sushi) dominates, it means it’s popular but also competitive. Alternatively, a less represented but still popular category may offer a niche opportunity with reduced competition.

**Decision-Making:** Choose a cuisine that balances popularity and competition levels. If aiming for a popular category, consider differentiating with unique offerings or exceptional service.

# 2. Best Locations for Restaurants Based on Average Ratings

```python
query = """
SELECT TOP 5 address, AVG(stars) as avg_rating
FROM BUSINESS
WHERE city = 'Philadelphia' AND categories LIKE '%Restaurants%'
GROUP BY address
ORDER BY avg_rating DESC;
"""

# Execute the query and store the results in a DataFrame
df_popular_categories = pd.read_sql(query, engine)
print(df_popular_categories)
```

**Insight:** Locations with high average ratings suggest areas where customers are likely more satisfied with dining experiences, possibly due to affluent clientele, better service standards, or vibrant surroundings.

**Decision-Making:** Target high-rating areas for higher foot traffic and brand prestige, or select an up-and-coming area with fewer restaurants but a promising rating trend to capture new market potential.

# 3. Average Rating by Cuisine Type in the City

```python
query = """
SELECT categories, AVG(stars) as avg_rating
FROM BUSINESS
WHERE city = 'Philadelphia' AND categories LIKE '%Restaurants%'
GROUP BY categories
ORDER BY avg_rating DESC;
"""

# Execute the query and store the results in a DataFrame
df_rate = pd.read_sql(query, engine)
print(df_rate)
```

**Insight:** Different cuisine types may have varying customer satisfaction levels. A consistently high rating in a particular category (e.g., Mediterranean) suggests strong customer appreciation for that cuisine, even if it’s less common.

**Decision-Making:** If a cuisine type with high ratings also has lower competition, it could be a smart choice. Alternatively, a cuisine with low ratings might indicate unmet customer expectations, so you can focus on solving those issues (e.g., healthier menu options, authentic preparation).

# 4.  Good experience of customers Based on Tip

```python
query = """
SELECT 
    COUNT(*) AS checkin_count
FROM TIP
WHERE text LIKE '%good%'
ORDER BY checkin_count DESC;
"""

# Execute the query and store the results in a DataFrame
df_demand = pd.read_sql(query, engine)
print(df_demand)
```

**Insight:** Knowing when customers give good comments helps with staffing, marketing, and promotional planning.

**Decision-Making:** Design attractive customer service.

# 5. Top 5 Restaurants with the Most Reviews

```python
query = """
SELECT TOP 5 name, review_count
FROM BUSINESS
WHERE city = 'Philadelphia' AND categories LIKE '%Restaurants%'
ORDER BY review_count DESC;
"""

# Execute the query and store the results in a DataFrame
df_mostreview = pd.read_sql(query, engine)
print(df_mostreview)
```

**Insight:** Highly reviewed restaurants have either a strong brand presence or are located in high-traffic areas. Examining these locations and restaurants could reveal what draws customers, such as ambiance, menu variety, or pricing.

**Decision-Making:** Investigate these top restaurants for potential lessons on what attracts diners in the city. Emulating aspects like ambiance, portion sizes, or engagement strategies could help in building a similar customer base.

# 6. Average Rating of Restaurants by Zip Code

```python
query = """
SELECT postal_code, AVG(stars) as avg_rating
FROM BUSINESS
WHERE city = 'Philadelphia' AND categories LIKE '%Restaurants%'
GROUP BY postal_code
ORDER BY avg_rating DESC;
"""

# Execute the query and store the results in a DataFrame
df_ratezip = pd.read_sql(query, engine)
print(df_ratezip)
```

**Insight:** This metric can show which neighborhoods have higher average restaurant ratings, often reflecting both customer satisfaction and demographic factors (e.g., disposable income or lifestyle preferences).

**Decision-Making:** Consider setting up in a high-rated zip code to align with established customer expectations. Alternatively, choose a lower-rated area if you believe you can exceed current standards and attract new customers by standing out.

# 7. Customer Feedback on Service Quality (Based on Keywords)

```python
query = """
SELECT review_id, text
FROM REVIEW
WHERE business_id IN (SELECT business_id FROM BUSINESS WHERE city = 'Philadelphia')
AND text LIKE '%service%';
"""

# Execute the query and store the results in a DataFrame
df_feedback = pd.read_sql(query, engine)
print(df_feedback)
```

**Insight:** Analyzing reviews mentioning "service" reveals common complaints or praise, helping identify what customers value most in service quality (e.g., friendliness, speed, knowledgeability).

**Decision-Making:** Use this feedback to shape your hiring, training, and customer service policies. If “friendly” is a recurring positive term, emphasize warmth in service; if “slow” is a frequent negative, optimize for speed.

# 8. Identify Gaps in Cuisine Offerings

```python
query = """
SELECT categories, COUNT(business_id) AS num
FROM BUSINESS
WHERE city = 'Philadelphia' AND categories LIKE '%Restaurants%'
GROUP BY categories
HAVING COUNT(business_id) < 5
ORDER BY num;
"""

# Execute the query and store the results in a DataFrame
df_gaps = pd.read_sql(query, engine)
print(df_gaps)
```

**Insight:** Categories with few restaurants represent potential niches, especially if there’s evidence of demand or if those few restaurants have high ratings.

**Decision-Making:** Consider introducing an underrepresented cuisine in the market. For example, if there are few Ethiopian or Brazilian restaurants but a growing demand, it may be a profitable opportunity with minimal competition.

# 9. Analyze Social Media Engagement with Restaurants

```python
query = """
SELECT business_id, COUNT(*) AS social_media_mentions
FROM REVIEW
WHERE text LIKE '%Instagram%' OR text LIKE '%Facebook%' OR text LIKE '%Twitter%'
AND business_id IN (SELECT business_id FROM BUSINESS WHERE city = 'Philadelphia')
GROUP BY business_id
ORDER BY social_media_mentions DESC;
"""

# Execute the query and store the results in a DataFrame
df_social = pd.read_sql(query, engine)
print(df_social)
```

**Insight:** Identifies restaurants with high social media mentions, indicating strong customer engagement and successful marketing tactics that resonate locally.

**Decision-Making:** Analyze popular restaurants’ social media strategies (e.g., influencer partnerships, photo-worthy dishes) and replicate effective tactics. Create visually appealing dishes and experiences to encourage sharing, and focus marketing efforts on platforms favored by your target audience.

# 10. Top Attributes Valued by Customers in Highly Rated Restaurants

```python
query = """
SELECT attributes, COUNT(*) as frequency
FROM BUSINESS
WHERE city = 'Philadelphia' AND stars >= 4
GROUP BY attributes
ORDER BY frequency DESC;
"""

# Execute the query and store the results in a DataFrame
df_attributes = pd.read_sql(query, engine)
print(df_attributes)
```

**Insight:** By identifying frequently mentioned attributes in positive reviews (e.g., “outdoor seating,” “organic ingredients,” “locally sourced”), you can understand which elements resonate most with customers.

**Decision-Making:** Incorporate the top-rated attributes into your restaurant concept. For example, if “outdoor seating” is popular, prioritize a layout with patio space, or if “organic” is a valued term, integrate organic options into your menu.
